```r
library(tidyverse)

# Load the data from Excel, skipping the first row and using the second row as the header
data <- readxl::read_excel("Data/UntidyData_Badly-Structured-Sales-Dat.xlsx", skip = 1)

# Rename the first column to "Order ID" for clarity
colnames(data)[1] <- "Order ID"

# Remove columns or rows with "Total" in their names
data <- data %>% select(-contains("Total"))
data <- data[!grepl("Total", data$`Order ID`), ]

# Pivot the data to long format
data_long <- data %>%
  pivot_longer(
    cols = -`Order ID`, 
    names_to = c("Segment", "Ship Mode"), 
    names_sep = " ", 
    values_to = "Sales"
  ) %>%
  # Filter out NA Sales values to clean up the dataset
  filter(!is.na(Sales))

# View the resulting tibble for verification
print(data_long)
```
