### Data Mining: Homework 1 - Python Implementation

#### Question 1:

Download the excel file ”UntidyData Badly-Structured-Sales-Dat.csv” from Teams and re- arrange this data into the correct four columns. There has been a mix of rows and columns everywhere. Also, watch out for Grand Totals and Sub Totals, you do not need those in tidy data.



```python
import pandas as pd

data = pd.read_excel('Data/UntidyData_Badly-Structured-Sales-Dat.xlsx')
data.head()
```

#### 1. Dropping Total Columns

```python
# Drop the 'Total' columns
data = data.drop(columns=['Consumer Total', 'Corporate Total', 'Home Office Total'])

data.head()
```

#### 2. Creating New Columns


```python
order_type = sorted(['Consumer', 'Corporate', 'Home Office'] * 4)
order_type.insert(0, 'Segment')

# Combine first row and newly created columns
column_names = [f'{x}|{y}' for x, y in zip(order_type, data.iloc[0].values)]
column_names
```

```python
# set column names
data.columns = column_names
```

```python
data.head()
```

#### Drop the row column



```python
data = data.iloc[1:-1, :]

data.head()
```

#### Pivot Long. 

In Pandas this is equivalent to melt function. You can also use the pivot function.

```python
data_long = pd.melt(data, id_vars=column_names[0], var_name='OrderType', value_name='Value')
data_long.head()
```

#### Expand the Dataset bt Splitting OrderType

```python
data_long[['OrderType', 'Class']] = data_long['OrderType'].str.split('|', expand=True)

data_long.head()
```

#### Drop NAs



```python
data_long = data_long.dropna()
data_long.head()
```

```python
data_long.columns
```

#### Reset Column Names


```python
data_long = data_long[['Segment|Ship Mode>>', 'OrderType', 'Class', 'Value'] ]
data_long.columns = ['OrderID', 'OrderType', 'Class', 'OrderValue']
```

```python
data_long.head()
```

### Question 2.


Download the excel file ”UntidyInvoices-with-Merged-Categories-and-Merged-Amounts.csv” from Teams and tidy it. Because a single transaction (identified with an order id ) has multiple items purchased, who ever captured this data decided to create a single row for each order, thereby lumping the different items purchased and the amounts together into 2 fields respectively. The better thing to do is to let each item purchased be on a single row with the amount. It is better to repeat the Order IDs on different rows than lumping up amounts in a single cell. As we will be analyzing items bought and amounts a lot, we need them separated into rows.


```python
invoices = pd.read_excel('Data/UntidyInvoices-with-Merged-Categories-and-Merged-Amounts.xlsx')
invoices
```

#### 2.1. Implementing Pivot_Longer for Categories Only


```python
categories = invoices['Category'].str.split('|', expand=True)
categories
```

```python
category_long = pd.melt(pd.concat([invoices[['Order ID']], categories], axis=1), id_vars=['Order ID'])
category_long
```

#### 2.2. Implementing Pivot_Longer for Amounts Only

```python
amount = invoices['Amount'].str.split('|', expand=True)

amount_long = pd.melt(pd.concat([invoices[['Order ID']], amount], axis=1), id_vars=['Order ID'])
amount_long
```

#### 2.3. Joining the Amount and Categories Long Formats

This combines both tables to create a final long table.

```python
final_df = pd.merge(category_long, amount_long, on=['Order ID', 'variable'], how='outer').dropna().drop(['variable'], axis=1)
final_df.columns = ['OrderID', 'Category', 'Amount']
```
